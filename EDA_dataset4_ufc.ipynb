```python
# ==========================================
# CELDA 3 — CARGA Y EXPLORACION INICIAL
# ==========================================

# Cargar dataset principal
df = pd.read_csv("/root/.cache/kagglehub/datasets/aminealibi/ufc-fights-fighters-and-events-dataset/versions/87/data/Fights.csv")

print("==========================================")
print("DATASET: UFC Fights")
print("==========================================")
print("\n--- Shape ---")
print("Filas:    " + str(df.shape[0]))
print("Columnas: " + str(df.shape[1]))

print("\n--- Columnas ---")
for col in df.columns:
    print("  - " + col)

print("\n--- Primeras 5 filas ---")
display(df.head())

print("\n--- Tipos de datos ---")
print(df.dtypes)

print("\n--- Estadísticas descriptivas ---")
display(df.describe())
```

```python
# ==========================================
# CELDA 4 — NULOS, DUPLICADOS Y TRADUCCION
# ==========================================

# Nulos
print("--- Valores nulos por columna ---")
nulos = df.isnull().sum()
porcentaje = (nulos / len(df)) * 100
df_nulos = pd.DataFrame({'Nulos': nulos, 'Porcentaje': porcentaje})
df_nulos = df_nulos[df_nulos['Nulos'] > 0]
if len(df_nulos) == 0:
    print("Sin valores nulos.")
else:
    display(df_nulos)

# Duplicados
print("\n--- Filas duplicadas ---")
duplicados = df.duplicated().sum()
print("Total: " + str(duplicados))
if duplicados > 0:
    df = df.drop_duplicates()
    print("Filas después de eliminar: " + str(df.shape[0]))

# Traduccion de columnas
df = df.rename(columns={
    'Fight_Id':         'ID_Pelea',
    'Fighter_Id_1':     'ID_Peleador_1',
    'Fighter_Id_2':     'ID_Peleador_2',
    'Fighter_1':        'Peleador_1',
    'Fighter_2':        'Peleador_2',
    'KD_1':             'Knockdowns_1',
    'KD_2':             'Knockdowns_2',
    'STR_1':            'Golpes_1',
    'STR_2':            'Golpes_2',
    'TD_1':             'Derribos_1',
    'TD_2':             'Derribos_2',
    'SUB_1':            'Sumisiones_1',
    'SUB_2':            'Sumisiones_2',
    'Weight_Class':     'Categoria_Peso',
    'Method':           'Metodo_Victoria',
    'Round':            'Ronda',
    'Fight_Time':       'Tiempo_Pelea',
    'Event_Id':         'ID_Evento',
    'Result_1':         'Resultado_1',
    'Result_2':         'Resultado_2',
    'Time Format':      'Formato_Tiempo',
    'Referee':          'Arbitro',
    'Method Details':   'Detalle_Metodo',
    'Sig. Str. %_1':    'Pct_Golpes_Significativos_1',
    'Sig. Str. %_2':    'Pct_Golpes_Significativos_2',
    'Sub. Att_1':       'Intentos_Sumision_1',
    'Sub. Att_2':       'Intentos_Sumision_2',
    'Rev._1':           'Reversales_1',
    'Rev._2':           'Reversales_2',
    'Ctrl_1':           'Control_Suelo_1',
    'Ctrl_2':           'Control_Suelo_2',
    'Head_%_1':         'Pct_Cabeza_1',
    'Head_%_2':         'Pct_Cabeza_2',
    'Body_%_1':         'Pct_Cuerpo_1',
    'Body_%_2':         'Pct_Cuerpo_2',
    'Leg_%_1':          'Pct_Piernas_1',
    'Leg_%_2':          'Pct_Piernas_2',
    'Distance_%_1':     'Pct_Distancia_1',
    'Distance_%_2':     'Pct_Distancia_2',
    'Clinch_%_1':       'Pct_Clinch_1',
    'Clinch_%_2':       'Pct_Clinch_2',
    'Ground_%_1':       'Pct_Suelo_1',
    'Ground_%_2':       'Pct_Suelo_2',
    'Total Str._%_1':   'Pct_Total_Golpes_1',
    'Total Str._%_2':   'Pct_Total_Golpes_2',
    'Sig. Str._%_1':    'Pct_Sig_Golpes_1',
    'Sig. Str._%_2':    'Pct_Sig_Golpes_2',
})

# Traducir valores categoricos
df['Resultado_1'] = df['Resultado_1'].map({'W': 'Victoria', 'L': 'Derrota', 'D': 'Empate', 'NC': 'Sin_Resultado'})
df['Resultado_2'] = df['Resultado_2'].map({'W': 'Victoria', 'L': 'Derrota', 'D': 'Empate', 'NC': 'Sin_Resultado'})

print("\n--- Columnas traducidas ---")
print(df.columns.tolist())
print("\n--- Valores únicos Resultado_1 ---")
print(df['Resultado_1'].value_counts())
print("\n--- Valores únicos Metodo_Victoria ---")
print(df['Metodo_Victoria'].value_counts())
print("\n--- Valores únicos Categoria_Peso ---")
print(df['Categoria_Peso'].value_counts())
```

```python
# ==========================================
# CELDA 5 — OUTLIERS Y VALORES INAPROPIADOS
# ==========================================

# Columnas numericas clave
columnas_numericas = ['Knockdowns_1', 'Knockdowns_2', 'Golpes_1', 'Golpes_2',
                      'Derribos_1', 'Derribos_2', 'Sumisiones_1', 'Sumisiones_2', 'Ronda']

# Outliers por IQR
print("--- Outliers por método IQR ---")
resumen = []
for col in columnas_numericas:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lim_inf = Q1 - 1.5 * IQR
    lim_sup = Q3 + 1.5 * IQR
    n_out = df[(df[col] < lim_inf) | (df[col] > lim_sup)].shape[0]
    resumen.append({
        'Columna': col,
        'Q1': round(Q1, 2),
        'Q3': round(Q3, 2),
        'Limite_Inferior': round(lim_inf, 2),
        'Limite_Superior': round(lim_sup, 2),
        'Outliers': n_out
    })
display(pd.DataFrame(resumen))

# Distribucion del target
print("\n--- Distribución del Target (Resultado_1) ---")
print(df['Resultado_1'].value_counts())
print("\nNota: Para clasificación binaria se usarán solo Victoria y Derrota")
df_modelo = df[df['Resultado_1'].isin(['Victoria', 'Derrota'])]
print("Filas útiles para el modelo: " + str(df_modelo.shape[0]))

# Categorias de peso
print("\n--- Traducción Categoría de Peso ---")
df['Categoria_Peso'] = df['Categoria_Peso'].map({
    'Lightweight':              'Peso_Ligero',
    'Welterweight':             'Peso_Welter',
    'Middleweight':             'Peso_Medio',
    'Featherweight':            'Peso_Pluma',
    'Bantamweight':             'Peso_Gallo',
    'Heavyweight':              'Peso_Pesado',
    'Light Heavyweight':        'Semipesado',
    'Flyweight':                'Peso_Mosca',
    "Women's Strawweight":      'Femenino_Paja',
    "Women's Flyweight":        'Femenino_Mosca',
    "Women's Bantamweight":     'Femenino_Gallo',
    'Open Weight':              'Peso_Abierto',
    'Catch Weight':             'Peso_Acordado',
    "Women's Featherweight":    'Femenino_Pluma',
    'Super Heavyweight':        'Superpesado'
})
print(df['Categoria_Peso'].value_counts())
```

```python
# ==========================================
# CELDA 6 — VISUALIZACIONES
# ==========================================

fig, axes = plt.subplots(2, 3, figsize=(16, 10))
axes = axes.flatten()

# 1. Distribución del Resultado
resultado_conteo = df['Resultado_1'].value_counts()
axes[0].bar(resultado_conteo.index, resultado_conteo.values,
            color=['steelblue', 'coral', 'gray', 'lightgray'])
axes[0].set_title('Distribución del Resultado (Peleador 1)')
axes[0].set_ylabel('Cantidad de peleas')
for i, v in enumerate(resultado_conteo.values):
    axes[0].text(i, v + 30, str(v), ha='center', fontweight='bold')

# 2. Victorias por Categoría de Peso
victorias_peso = df[df['Resultado_1'] == 'Victoria']['Categoria_Peso'].value_counts().head(10)
axes[1].barh(victorias_peso.index, victorias_peso.values, color='steelblue')
axes[1].set_title('Top 10 Categorías con más Victorias')
axes[1].set_xlabel('Cantidad')

# 3. Método de Victoria
metodo_top = df['Metodo_Victoria'].value_counts().head(8)
axes[2].barh(metodo_top.index, metodo_top.values, color='seagreen')
axes[2].set_title('Top 8 Métodos de Victoria')
axes[2].set_xlabel('Cantidad')

# 4. Distribución de Golpes (Peleador 1 vs 2)
axes[3].hist(df['Golpes_1'], bins=40, alpha=0.6, color='steelblue', label='Peleador 1')
axes[3].hist(df['Golpes_2'], bins=40, alpha=0.6, color='coral', label='Peleador 2')
axes[3].set_title('Distribución de Golpes por Peleador')
axes[3].set_xlabel('Golpes totales')
axes[3].set_ylabel('Frecuencia')
axes[3].legend()

# 5. Knockdowns por Resultado
df_bin = df[df['Resultado_1'].isin(['Victoria', 'Derrota'])]
kd_resultado = df_bin.groupby('Resultado_1')['Knockdowns_1'].mean()
axes[4].bar(kd_resultado.index, kd_resultado.values, color=['steelblue', 'coral'])
axes[4].set_title('Promedio de Knockdowns por Resultado')
axes[4].set_ylabel('Knockdowns promedio')
for i, v in enumerate(kd_resultado.values):
    axes[4].text(i, v + 0.005, str(round(v, 3)), ha='center', fontweight='bold')

# 6. Ronda en que terminan las peleas
ronda_conteo = df['Ronda'].value_counts().sort_index()
axes[5].bar(ronda_conteo.index, ronda_conteo.values, color='purple')
axes[5].set_title('Ronda en que Termina la Pelea')
axes[5].set_xlabel('Ronda')
axes[5].set_ylabel('Cantidad de peleas')
for i, (ronda, v) in enumerate(ronda_conteo.items()):
    axes[5].text(ronda, v + 20, str(v), ha='center', fontweight='bold')

plt.suptitle('Análisis Visual — Dataset UFC Fights', fontsize=14, y=1.02)
plt.tight_layout()
plt.show()
```

```python
# ==========================================
# CELDA 7 — MAPA DE CORRELACION
# ==========================================

# Solo columnas numericas relevantes
columnas_corr = ['Knockdowns_1', 'Knockdowns_2', 'Golpes_1', 'Golpes_2',
                 'Derribos_1', 'Derribos_2', 'Sumisiones_1', 'Sumisiones_2',
                 'Ronda', 'Pct_Golpes_Significativos_1', 'Pct_Golpes_Significativos_2',
                 'Control_Suelo_1', 'Control_Suelo_2']

plt.figure(figsize=(12, 9))
sns.heatmap(
    df[columnas_corr].corr(),
    annot=True,
    fmt='.2f',
    cmap='coolwarm',
    linewidths=0.5,
    vmin=-1,
    vmax=1
)
plt.title('Mapa de Correlación — Dataset UFC Fights')
plt.tight_layout()
plt.show()
```

```python
# ==========================================
# CELDA 8 — METODO DE VICTORIA + CORRELACION LUCHA
# ==========================================

# Simplificar metodo de victoria en 3 grupos
def clasificar_metodo(metodo):
    if pd.isna(metodo):
        return 'Otro'
    metodo = str(metodo).upper()
    if 'KO' in metodo or 'TKO' in metodo:
        return 'KO/TKO (Golpes)'
    elif 'SUB' in metodo:
        return 'Sumision (Lucha)'
    elif 'DEC' in metodo:
        return 'Decision (Puntos)'
    else:
        return 'Otro'

df['Tipo_Victoria'] = df['Metodo_Victoria'].apply(clasificar_metodo)

# Top 8 categorias de peso
top_pesos = df['Categoria_Peso'].value_counts().head(8).index
df_top = df[df['Categoria_Peso'].isin(top_pesos)]

# Tabla cruzada
tabla = pd.crosstab(df_top['Categoria_Peso'], df_top['Tipo_Victoria'], normalize='index') * 100
tabla = tabla.round(1)
print("--- % de cada tipo de victoria por categoría de peso ---")
display(tabla)

# Grafico 1 — metodo por categoria de peso
fig, axes = plt.subplots(1, 2, figsize=(18, 6))

tabla_plot = tabla[['KO/TKO (Golpes)', 'Sumision (Lucha)', 'Decision (Puntos)']]
tabla_plot.plot(kind='bar', ax=axes[0], colormap='Set2', edgecolor='white')
axes[0].set_title('Método de Victoria por Categoría de Peso (%)')
axes[0].set_ylabel('Porcentaje (%)')
axes[0].set_xlabel('Categoría de Peso')
axes[0].tick_params(axis='x', rotation=45)
axes[0].legend(title='Tipo de Victoria')

# Grafico 2 — correlacion variables de lucha
columnas_lucha = [
    'Derribos_1', 'Derribos_2',
    'Sumisiones_1', 'Sumisiones_2',
    'Control_Suelo_1', 'Control_Suelo_2',
    'Intentos_Sumision_1', 'Intentos_Sumision_2',
    'Reversales_1', 'Reversales_2'
]
sns.heatmap(
    df[columnas_lucha].corr(),
    annot=True,
    fmt='.2f',
    cmap='coolwarm',
    linewidths=0.5,
    vmin=-1,
    vmax=1,
    ax=axes[1]
)
axes[1].set_title('Correlación entre Variables de Lucha')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
# ==========================================
# CELDA 8b — GOLPES VS LUCHA: ¿QUIEN GANA MAS?
# ==========================================

df_bin = df[df['Resultado_1'].isin(['Victoria', 'Derrota'])].copy()

# Contar victorias por tipo de victoria
victorias_tipo = df_bin[df_bin['Resultado_1'] == 'Victoria']['Tipo_Victoria'].value_counts()
total = victorias_tipo.sum()

print("--- Victorias por tipo de estrategia ---")
for tipo, cantidad in victorias_tipo.items():
    pct = round(cantidad / total * 100, 1)
    print(f"  {tipo:<25}: {cantidad} victorias ({pct}%)")

# Grafico
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Torta
axes[0].pie(victorias_tipo.values,
            labels=victorias_tipo.index,
            autopct='%1.1f%%',
            colors=['coral', 'steelblue', 'seagreen', 'gray'],
            startangle=90)
axes[0].set_title('¿Cómo se ganan las peleas en UFC?')

# Promedio de golpes y derribos por resultado
metricas = df_bin.groupby('Resultado_1')[['Golpes_1', 'Derribos_1',
                                          'Control_Suelo_1', 'Knockdowns_1']].mean().round(2)
metricas.plot(kind='bar', ax=axes[1], colormap='Set2', edgecolor='white')
axes[1].set_title('Promedio de Estadísticas por Resultado')
axes[1].set_ylabel('Promedio')
axes[1].tick_params(axis='x', rotation=0)
axes[1].legend(title='Estadística')

plt.tight_layout()
plt.show()

print("\n--- Promedio de estadísticas: Victoria vs Derrota ---")
display(metricas)
```

```python
# ==========================================
# CELDA 8c — TIPO DE GOLPE + SUMISIONES VS CONTROL PISO
# ==========================================

fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# 1. Promedio de golpes por zona (cabeza, cuerpo, piernas) segun resultado
zonas_victoria = df_bin[df_bin['Resultado_1'] == 'Victoria'][
    ['Pct_Cabeza_1', 'Pct_Cuerpo_1', 'Pct_Piernas_1']].mean()
zonas_derrota = df_bin[df_bin['Resultado_1'] == 'Derrota'][
    ['Pct_Cabeza_1', 'Pct_Cuerpo_1', 'Pct_Piernas_1']].mean()

x = ['Cabeza', 'Cuerpo', 'Piernas']
x_pos = range(len(x))
width = 0.35

axes[0, 0].bar([p - width/2 for p in x_pos], zonas_victoria.values,
               width=width, color='steelblue', label='Victoria')
axes[0, 0].bar([p + width/2 for p in x_pos], zonas_derrota.values,
               width=width, color='coral', label='Derrota')
axes[0, 0].set_xticks(list(x_pos))
axes[0, 0].set_xticklabels(x)
axes[0, 0].set_title('Distribución de Golpes por Zona (Victoria vs Derrota)')
axes[0, 0].set_ylabel('Porcentaje promedio')
axes[0, 0].legend()

# 2. Golpes por distancia (distancia, clinch, suelo) segun resultado
distancias_victoria = df_bin[df_bin['Resultado_1'] == 'Victoria'][
    ['Pct_Distancia_1', 'Pct_Clinch_1', 'Pct_Suelo_1']].mean()
distancias_derrota = df_bin[df_bin['Resultado_1'] == 'Derrota'][
    ['Pct_Distancia_1', 'Pct_Clinch_1', 'Pct_Suelo_1']].mean()

x2 = ['Distancia', 'Clinch', 'Suelo']
x_pos2 = range(len(x2))

axes[0, 1].bar([p - width/2 for p in x_pos2], distancias_victoria.values,
               width=width, color='steelblue', label='Victoria')
axes[0, 1].bar([p + width/2 for p in x_pos2], distancias_derrota.values,
               width=width, color='coral', label='Derrota')
axes[0, 1].set_xticks(list(x_pos2))
axes[0, 1].set_xticklabels(x2)
axes[0, 1].set_title('Golpes por Posición (Victoria vs Derrota)')
axes[0, 1].set_ylabel('Porcentaje promedio')
axes[0, 1].legend()

# 3. Scatter: Control de Suelo vs Sumisiones — color por resultado
colores = df_bin['Resultado_1'].map({'Victoria': 'steelblue', 'Derrota': 'coral'})
axes[1, 0].scatter(df_bin['Control_Suelo_1'], df_bin['Sumisiones_1'],
                   c=colores, alpha=0.3, s=15)
axes[1, 0].set_title('Control de Suelo vs Sumisiones\n(Azul=Victoria, Rojo=Derrota)')
axes[1, 0].set_xlabel('Control de Suelo (seg)')
axes[1, 0].set_ylabel('Sumisiones')

# 4. Promedio de Control Suelo e Intentos Sumision por resultado
lucha_stats = df_bin.groupby('Resultado_1')[
    ['Control_Suelo_1', 'Sumisiones_1', 'Intentos_Sumision_1', 'Derribos_1']].mean().round(3)

lucha_stats.plot(kind='bar', ax=axes[1, 1], colormap='coolwarm', edgecolor='white')
axes[1, 1].set_title('Estadísticas de Lucha: Victoria vs Derrota')
axes[1, 1].set_ylabel('Promedio')
axes[1, 1].tick_params(axis='x', rotation=0)
axes[1, 1].legend(title='Estadística', fontsize=8)

plt.suptitle('Análisis: Tipo de Golpe y Lucha en el Suelo', fontsize=14, y=1.02)
plt.tight_layout()
plt.show()

# Tabla resumen lucha
print("\n--- Promedio estadísticas de lucha por resultado ---")
display(lucha_stats)

# Correlacion sumisiones vs control suelo
corr = df_bin[['Control_Suelo_1', 'Sumisiones_1', 'Intentos_Sumision_1',
               'Derribos_1', 'Reversales_1']].corr()
print("\n--- Correlación: Sumisiones vs Control de Suelo ---")
display(corr)
```

```python
# ==========================================
# CELDA 8d — QUE GENERA UNA VICTORIA POR DECISION
# ==========================================

# Separar por tipo de victoria
df_dec = df_bin[df_bin['Tipo_Victoria'] == 'Decision (Puntos)'].copy()
df_ko  = df_bin[df_bin['Tipo_Victoria'] == 'KO/TKO (Golpes)'].copy()
df_sub = df_bin[df_bin['Tipo_Victoria'] == 'Sumision (Lucha)'].copy()

# Promedio de estadisticas por tipo de victoria — solo ganadores
stats_cols = ['Golpes_1', 'Knockdowns_1', 'Derribos_1',
              'Control_Suelo_1', 'Sumisiones_1', 'Pct_Cabeza_1',
              'Pct_Piernas_1', 'Pct_Distancia_1', 'Pct_Suelo_1']

resumen = pd.DataFrame({
    'Decision':  df_dec[df_dec['Resultado_1'] == 'Victoria'][stats_cols].mean(),
    'KO/TKO':    df_ko[df_ko['Resultado_1']   == 'Victoria'][stats_cols].mean(),
    'Sumision':  df_sub[df_sub['Resultado_1'] == 'Victoria'][stats_cols].mean(),
}).round(3)

print("--- Estadísticas promedio del GANADOR por tipo de victoria ---")
display(resumen)

# Grafico
resumen.T.plot(kind='bar', figsize=(14, 6), colormap='Set2', edgecolor='white')
plt.title('¿Qué hace el ganador según cómo termina la pelea?')
plt.ylabel('Promedio')
plt.xticks(rotation=0)
plt.legend(title='Estadística', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

¿QUE HACE EL GANADOR SEGUN EL METODO?

  DECISION (puntos):
    61.6 golpes promedio — volumen alto, no potencia
    282 segundos de control de suelo — el más alto
    2.2 derribos — lleva la pelea al suelo para puntuar
    Golpea más las piernas (76.5%) — patadas bajas
    para acumular puntos en los jueces

  KO/TKO (golpes):
    Solo 32 golpes pero 0.77 knockdowns — menos pero
    más poderosos. Busca el nocaut, no el volumen.
    51.3% de golpes a la cabeza — objetivo claro
    Poco control de suelo — prefiere pelear de pie

  SUMISION (lucha):
    1.5 sumisiones promedio vs 0.4 en decisión
    157 segundos de control de suelo
    El peleador que somete derriba (~1.4 veces) y
    controla el suelo para buscar la rendición

HALLAZGO CLAVE — CONTROL DE SUELO:
  El ganador promedia 193 seg de control de suelo
  vs 84 seg del perdedor — más del doble.
  Es el predictor más fuerte de victoria general,
  independiente del método con que termine la pelea.

CORRELACIONES CLAVE:
  Derribos ↔ Control_Suelo: 0.72
  Quien derriba, controla. Quien controla, gana.

  Golpes_1 ↔ Golpes_2: 0.66
  Peleas activas generan daño para ambos lados.

  Sumisiones ↔ Intentos_Sumision: 1.00
  Variable duplicada — se eliminará en el modelado.
